```python
from sklearn.datasets import load_iris
iris = load_iris()
```

###### iris = load_iris(): This line loads the Iris dataset using the load_iris function and assigns it to the variable iris.

```python
import pandas as pd
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

###### This line creates a Pandas DataFrame from the Iris dataset and assigns it to the variable df.
###### The data attribute of the iris object contains the feature values, and the feature_names attributes contains the names of the features.


```python
df['target'] = iris.target

```

###### df['target'] = iris.target:This line adds a new column to the DataFrame called target and assigns it the values from the target attribute of the iris object. The target attribute contains the class labels for the Iris dataset.



##### Step 3: Create basic line plot and scatter plot with Matplotlib



```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))
### This line creates a new figure with a specified size (8x6 inches) using the figure function from matplotlib.pyplot.
plt.plot(df['sepal length (cm)'], label='Sepal Length')

###  plt.plot(df['sepal length (cm)'], label='Sepal Length'): 
# This line creates a line plot of the sepal length (cm) column from the DataFrame using the plot function from matplotlib.pyplot. 
# The label parameter specifies the label for the plot.
plt.plot(df['sepal width (cm)'], label='Sepal Width')
#### plt.plot(df['sepal width (cm)'], label='Sepal Width'): 
# This line creates another line plot of the sepal width (cm) column from the DataFrame using the plot function from matplotlib.pyplot. 
# The label parameter specifies the label for the plot.
plt.legend()
## This line adds a legend to the plot using the legend function from matplotlib.pyplot.
plt.title('Sepal Length and Width')
## plt.title('Sepal Length and Width'): This line sets the title of the plot using the title function from matplotlib.pyplot.
plt.xlabel('Index')
## plt.xlabel('Index'): This line sets the label for the x-axis using the xlabel function from matplotlib.pyplot.
plt.ylabel('Value')
## plt.ylabel('Value'): This line sets the label for the y-axis using the ylabel function from matplotlib.pyplot.
plt.show()
##  plt.show(): This line displays the plot using the show function from matplotlib.pyplot.
```

##### This line creates a scatter plot of the sepal length (cm) column against the sepal width (cm) column from the DataFrame.
##### You can also add labels, title, and legend to the scatter plot as follows:


```python
plt.figure(figsize=(8, 6))
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'])
plt.title('Sepal Length vs. Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
## This code will create a scatter plot with labels and title, and display it using the showÂ function.
```

##### using histogram 

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(df['sepal length (cm)'], kde=True)
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

#### FOR BOXPLOT WE HAVE THE CODE BELOW 

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x='target', y='sepal width (cm)', data=df)
plt.title('Boxplot of Sepal Width by Target')
plt.xlabel('Target')
plt.ylabel('Sepal Width')
plt.show()
```
