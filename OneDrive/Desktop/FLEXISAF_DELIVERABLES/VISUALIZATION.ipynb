```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales_data = pd.read_csv('company_sales_data.csv')
sales_data
```

```python
#Exercise 1  Read Total profit of all months and show it using a line plot
plt.figure(figsize=(10, 6))
plt.plot(sales_data['month_number'], sales_data['total_profit'], marker='o')
plt.title('Total Profit of All Months')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

```python
#Exercise 2 Read data for Bathing soap and facewash of all months and display it using the Subplot
fig, ax = plt.subplots(2, figsize=(10, 6))

ax[0].plot(sales_data['month_number'], sales_data['bathingsoap'], marker='o')
ax[0].set_title('Bathing Soap Sales')
ax[0].set_xlabel('Month')
ax[0].set_ylabel('Sales')
ax[0].grid(True)

ax[1].plot(sales_data['month_number'], sales_data['facewash'], marker='o')
ax[1].set_title('facewash Sales')
ax[1].set_xlabel('Month')
ax[1].set_ylabel('Sales')
ax[1].grid(True)

plt.tight_layout()
plt.show()

```
